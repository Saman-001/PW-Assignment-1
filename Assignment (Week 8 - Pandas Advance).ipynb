```python
def indexing_df(df):
    new_indexing =  pd.RangeIndex(start=1,stop=len(df)*2,step=2)
    reindexed_df = df.set_index(new_indexing)
    return reindexed_df
df_1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
df_reindexed = indexing_df(df_1)

# Print the re-indexed DataFrame
print(df_reindexed)
```

### Question 10

```python


def calculate_moving_average(df):
    # Assuming 'Date' column is in datetime format, if not, convert it to datetime first
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Sort the DataFrame by 'Date' if it's not already sorted
    df.sort_values(by='Date', inplace=True)
    
    # Set 'Date' as the index for the rolling operation
    df.set_index('Date', inplace=True)
    
    # Calculate the moving average with a window of size 7
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    
    # Reset the index to revert to the original DataFrame structure
    df.reset_index(inplace=True)
    
    return df

# Example usage:
# Assuming df is your original DataFrame with 'Sales' and 'Date' columns
# Replace the following line with your actual DataFrame
df6 = pd.DataFrame({
    'Date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05', '2022-01-06', '2022-01-07'],
    'Sales': [10, 12, 15, 18, 20, 22, 25]
})

# Call the function to add the 'MovingAverage' column
df_with_moving_average = calculate_moving_average(df6)

# Print the DataFrame with the new 'MovingAverage' column
print(df_with_moving_average)
```

### Question 11

```python
import pandas as pd

def add_weekday_column(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df['Weekday'] = df['Date'].dt.strftime('%A')
    
    return df
df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']})
df_with_weekday = add_weekday_column(df)
print(df_with_weekday)
```

```python
import pandas as pd

def filter_date_range(df):
    df['Date'] = pd.to_datetime(df['Date'])
    start_date = '2023-01-01'
    end_date = '2023-01-31'
    filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
    return filtered_df

df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-15', '2023-01-30', '2023-02-05']})
filtered_df = filter_date_range(df)
print(filtered_df)
```
