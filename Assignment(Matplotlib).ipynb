## Question 3

The subplot() function in Matplotlib is used to create multiple plots within the same figure. It allows you to organize and display multiple plots in a grid-like structure. 

```python
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
# Line 1
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('Line 1')
```

```python
# Line 2
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('Line 2')
```

```python
# Line 3
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('Line 3')
```

```python
# Line 4
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')
```

## Question 4

A bar plot is a graphical representation of data where individual bars or columns represent different categories, and the height (or length in the case of horizontal bar plots) of each bar corresponds to the value of a variable. Bar plots are commonly used to compare and visualize the distribution or values of different categories in a dataset.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

## Question 5

A box plot, also known as a box-and-whisker plot, is a graphical representation that provides a summary of the distribution of a dataset. It displays key statistical measures, including the median, quartiles, and potential outliers. 

The box plot is particularly useful for identifying the central tendency, spread, and skewness of the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.figure(figsize=(8,5))
plt.boxplot([box1,box2],labels=['Box 1', 'Box 2'], patch_artist=True)
plt.title('Box Plot of Box1 and Box2')
plt.xlabel('Box')
plt.ylabel('Values')
```
