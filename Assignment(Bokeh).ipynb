## Question 2

glyphs are the visual markers used to represent data on a plot. Glyphs can take various forms, such as circles, lines, rectangles, and more

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure
p = figure(title="Bokeh Glyph Example", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add a circle glyph to the figure
circle_glyph = p.circle(x, y, size=10, color="navy", alpha=0.5)

# Add a line glyph to the figure
line_glyph = p.line(x, y, line_width=2, line_color="green", line_dash="dashed")

# Show the plot in the Jupyter Notebook (inline)

show(p)
```

## Question 3

```python
p = figure(title="Customized Bokeh Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add a circle glyph to the figure
circle_glyph = p.circle(x, y, size=10, color="navy", alpha=0.5, legend_label="Data Points")

# Customize the appearance of the plot
p.title.text_color = "darkblue"  # Set title text color
p.title.text_font_size = "16pt"  # Set title font size

p.xaxis.axis_label_text_color = "green"  # Set x-axis label text color
p.yaxis.axis_label_text_color = "red"    # Set y-axis label text color

p.xaxis.major_label_text_color = "purple"  # Set x-axis major label text color
p.yaxis.major_label_text_color = "orange"  # Set y-axis major label text color

p.legend.title = "Legend"  # Set legend title
p.legend.label_text_color = "black"  # Set legend label text color

# Show the plot in the Jupyter Notebook (inline)

show(p)
```

## Question 5

Flask:
1. Basic Workflow:

Flask is a lightweight web framework for Python.
To embed a Bokeh plot, you create a Flask app with a route that renders an HTML template.
Inside the route function, you generate a Bokeh plot using Bokeh's plotting functions.
The plot is then embedded into the HTML template using the components function from Bokeh.
2. Templating Engine:

Flask uses Jinja2 as its templating engine.
The render_template function is used to render an HTML template and pass data (in this case, the Bokeh plot components) to the template.
3. Running the App:

After defining the route and generating the Bokeh plot, you run the Flask app.
Accessing the specified route in a web browser will display the embedded Bokeh plot.
Django:
1. Django Project Structure:

Django is a high-level web framework that follows the Model-View-Controller (MVC) architecture.
Views handle the application logic and are responsible for processing requests.
2. Views in Django:

In Django, you create a view function to handle a specific URL pattern.
The view function generates the Bokeh plot and renders an HTML template.
3. URL Patterns:

URL patterns are defined in the urls.py file.
A URL pattern associates a URL with a specific view function.
4. Django Templates:

Django uses its own templating engine.
HTML templates are stored in the templates directory within the app.
5. Running the App:

After setting up views, URL patterns, and templates, you run the Django development server.
Visiting the specified URL will display the embedded Bokeh plot.
