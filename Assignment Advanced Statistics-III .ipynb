### Question 1

Estimation statistics involves estimating population parameters from sample data. Point estimates are single values approximating these parameters. Interval estimates provide ranges of values along with confidence levels. Point estimates offer simplicity but lack information on variability. Interval estimates offer a range of likely values, incorporating uncertainty. Both are essential for making informed decisions based on data.

### Question 2

```python
import scipy.stats as stats

def estimate_population_mean(sample_mean, sample_std, sample_size, confidence_level=0.95):
    # Calculate the margin of error using the t-distribution
    margin_of_error = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1) * (sample_std / (sample_size ** 0.5))
    # Calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return lower_bound, upper_bound
```

H1: μ_male > μ_female

### Question 5



```python
import scipy.stats as stats

def two_sample_t_test(sample1, sample2, alpha=0.05, alternative='two-sided'):
    # Perform two-sample t-test
    t_statistic, p_value = stats.ttest_ind(sample1, sample2)
    
    # Determine the alternative hypothesis based on the specified alternative parameter
    if alternative == 'two-sided':
        reject_null = (p_value < alpha)
        p_value *= 2  # For two-sided test, multiply p-value by 2
    elif alternative == 'greater':
        reject_null = (p_value < alpha) and (t_statistic > 0)
    elif alternative == 'less':
        reject_null = (p_value < alpha) and (t_statistic < 0)
    
    # Return test results
    return t_statistic, p_value, reject_null
```

### QUestion 7

The steps involved in hypothesis testing are as follows:-

1. State Hypotheses: Formulate null (H0) and alternative (H1) hypotheses.
2. Choose Significance Level: Select a significance level (α).
3. Select Test Statistic: Choose an appropriate test statistic.
4. Collect Data and Compute Test Statistic: Gather sample data and calculate the test statistic.
5. Determine Critical Region: Define the critical region.
6. Calculate P-value: Compute the probability of obtaining the test statistic under H0.
7. Make Decision: Compare the p-value to the significance level (α) and decide whether to reject H0.
8. Draw Conclusions: Interpret the results in the context of the research question.

### Question 8

The p-value, or probability value, is a measure used in hypothesis testing to determine the strength of evidence against the null hypothesis (H0). It represents the probability of obtaining the observed results (or more extreme results) when the null hypothesis is true.

The p-value measures the strength of evidence against the null hypothesis. A smaller p-value indicates stronger evidence to reject the null hypothesis, guiding decisions in hypothesis testing.

```python
import numpy as np
import scipy.stats as stats

def two_sample_t_test(sample1, sample2):
    # Calculate the test statistic and p-value using scipy's ttest_ind function
    t_statistic, p_value = stats.ttest_ind(sample1, sample2)
    return t_statistic, p_value

# Generate two random samples of equal size
sample_size = 30
np.random.seed(42)  # for reproducibility
sample1 = np.random.normal(loc=10, scale=2, size=sample_size)  # Example: Sample 1
sample2 = np.random.normal(loc=12, scale=2, size=sample_size)  # Example: Sample 2

# Perform the two-sample t-test
t_statistic, p_value = two_sample_t_test(sample1, sample2)

# Print the test results
print("Test Statistic:", t_statistic)
print("P-value:", p_value)
```

The t-statistic measures the difference between two sample means relative to the variability within each group. It's calculated as the difference in means divided by the standard error of the difference.

The formula is:

t = (X1-X2)/( ((s1)^2/n1) + ((s2)^2/n2)) )^1/2

where, X1 and X2 are the sample means, s1 and s2 are the sample std, n1 and n2 are the sample sizes
