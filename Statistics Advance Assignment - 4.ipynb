### Question 2

One Tailed test:
1. Tests for a specific direction of the effect (e.g., whether a parameter is greater than or less than a certain value).
2. The critical region for rejecting the null hypothesis is located in one tail of the distribution.
3. Used when the research hypothesis predicts the direction of the effect (e.g., testing if a new drug increases or decreases blood pressure compared to a placebo).

Two tailed test:
1. Tests for any difference without specifying the direction (e.g., whether a parameter is different from a certain value, either higher or lower).
2. The critical region for rejecting the null hypothesis is split between both tails of the distribution.
3. Used when the research hypothesis does not predict the direction of the effect (e.g., testing if a new drug has any effect on blood pressure, either increase or decrease, compared to a placebo).

### Question 3
Type 1 Error(False positive):
1. Occurs when the null hypothesis (H0) is true, but we incorrectly reject it.
e.g = A medical test for a new drug shows it is effective in reducing blood pressure when, in reality, the drug has no effect. Here, the null hypothesis (the drug has no effect) is true, but the test incorrectly concludes that the drug is effective.

Type 2 Error(False Negative):
1. Occurs when the null hypothesis (H0) is false, but we fail to reject it.
e.g = A study testing a new educational program fails to show an improvement in student performance, even though the program is actually effective. Here, the null hypothesis (the program has no effect) is false, but the test incorrectly concludes that the program has no effect.
